### 导入包

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')

from sklearn.preprocessing import StandardScaler

%matplotlib inline
%config InlineBackend.figure_format = 'svg'  # 图片输出为矢量图，保证清晰度

import warnings
warnings.filterwarnings('ignore')
```

### 导入数据

```python
data=pd.read_csv('./cancer-data.csv')
data=data.drop(['Unnamed: 0'],axis=1)
X=data.drop(['Group'],axis=1)
y=data['Group']
```

### 数据归一化

```python
column=X.columns
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X=sc.fit_transform(X)
X=pd.DataFrame(data=X,columns=column)
```

### 样本层面数据分析

```python
df=data
fig,ax = plt.subplots(3,3,figsize=(12,12))
for a in ax.flat:
    a.set_facecolor('white')
sns.distplot(df['Age'],ax=ax[0][0])
sns.histplot(data =df,x='Age',ax=ax[0][1],hue='Group',kde=True)
sns.boxplot(x=df['Group'],y=df['Age'],ax=ax[0][2])
sns.distplot(df['Gender'],ax=ax[1][0])
sns.histplot(data =df,x='Gender',ax=ax[1][1],hue='Group',kde=True)
sns.boxplot(x=df['Group'],y=df['Gender'],ax=ax[1][2])
sns.distplot(df['APG_0101010055P'],ax=ax[2][0])
sns.histplot(data =df,x='APG_0101010055P',ax=ax[2][1],hue='Group',kde=True)
sns.boxplot(x=df['Group'],y=df['APG_0101010055P'],ax=ax[2][2])
plt.savefig('C:/Users/睿/Desktop/ex1try/Sample based explanation.jpg', dpi=500, bbox_inches='tight',facecolor='white')
plt.show()
```

# 样本层面数据分析

```python
import time

# 记录开始时间
start_time = time.time()

df=data
fig,ax = plt.subplots(3,3,figsize=(12,12))
for a in ax.flat:
    a.set_facecolor('white')
    
sns.histplot(data =df,x='Age',ax=ax[0][0],hue='Group',kde=True)
sns.distplot(df['Age'],ax=ax[0][1])
sns.violinplot(x=df['Group'], y=df['Age'],ax=ax[0][2])
sns.histplot(data =df,x='Gender',ax=ax[1][0],hue='Group',kde=True)
sns.distplot(df['Gender'],ax=ax[1][1])
sns.violinplot(x=df['Group'],y=df['Gender'],ax=ax[1][2])
sns.histplot(data =df,x='APG_0101010009P',ax=ax[2][0],hue='Group',kde=True)
sns.distplot(df['APG_0101010009P'],ax=ax[2][1])
sns.violinplot(x=df['Group'],y=df['APG_0101010009P'],ax=ax[2][2])
plt.savefig('C:/Users/睿/Desktop/ex1try/Sample based explanation-2.jpg', dpi=500, bbox_inches='tight',facecolor='white')
plt.show()

# 记录结束时间
end_time = time.time()

# 计算并打印运行时间
print(f"运行时间: {end_time - start_time} 秒")
```
